```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from time import sleep
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
url = 'https://www.transtats.bts.gov/Data_Elements.aspx?Data=4'
driver.get(url)
```

```python
airports_df = pd.DataFrame()
```

```python
airports = []
airport_list = driver.find_element(By.ID,'AirportList')
a = Select(airport_list)
for i in a.options:
    airports.append(i.text)
```

```python
for i in range(2,5):
    airport_list = driver.find_element(By.ID,'AirportList')
    a = Select(airport_list)
    a.select_by_index(i)
    driver.implicitly_wait(10)

    submit= driver.find_element(By.ID,'Submit')
    driver.implicitly_wait(10)

    ActionChains(driver).move_to_element(submit).click(submit).perform()

    df = pd.read_html(driver.page_source)[0]
    df = df.iloc[49:]
    df.columns = ['Year','Month','Domestic','International','Total']
    df['Airport Name']=airports[i]
    print(df)
    airports_df=airports_df.append(df)
    driver.refresh()
    driver.implicitly_wait(10)
```
